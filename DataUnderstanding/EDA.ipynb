#### Value counts

```python
# Get the frequency count for each category in the 'Type' column
a = df["Type"].value_counts()

# Define labels (x-axis) and values (y-axis) for the bar plot
x = a.index[:3]  # Extract the first three labels ('L', 'M', 'H', assuming they are the most frequent)
y = a.values[:3]  # Extract the corresponding frequencies

# Create the bar plot
plt.figure(figsize=(4, 4))  # Set the figure size
pastel_dark_colors = ['#ff6961','#779ecb','#77dd77']
plt.bar(x, y, color=pastel_dark_colors, width=0.6) 

# Add title and labels for the axes
plt.title('Type Distribution', fontsize=14)
plt.xlabel('Type', fontsize=12)
plt.ylabel('Frequency', fontsize=12)

# Add labels with the values below each bar
for i in range(len(x)):
    plt.text(i, y[i] + 250, str(y[i]), ha='center', va='top', fontsize=10, color='black')  # Position text below the bars

# Display the plot
plt.tight_layout()  # Adjust layout to prevent overlapping
plt.show()
```
